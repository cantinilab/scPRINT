```python
import pandas as pd
```

```python
res = pd.read_csv('../../data/results_grnbench.tsv', sep="\t")
```

```python
res
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create the violin plot
plt.figure(figsize=(12, 8))
violin_plot = sns.violinplot(x='metrics', y='GRN (EPR)', data=res)

# Set the title and labels of the plot
violin_plot.set_title('Violin plot of GRN (EPR) across different metrics')
violin_plot.set_xlabel('Metrics')
violin_plot.set_ylabel('GRN (EPR)')

# Rotate the x labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Create the violin plot
plt.figure(figsize=(12, 8))
violin_plot = sns.violinplot(x='metrics', y='GRN (enrich %)', data=res)

# Set the title and labels of the plot
violin_plot.set_title('Violin plot of GRN TF target enrichment across different metrics')
violin_plot.set_xlabel('Metrics')
violin_plot.set_ylabel('target enrichment (%)')

# Rotate the x labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Count the number of GRN (celltype) that are 'Y' and calculate the percentage
total_counts = res.groupby('metrics').size()
grn_celltype_y_counts = res[res['GRN (celltype)'] == 'Y'].groupby('metrics').size()
grn_celltype_y_percentage = (grn_celltype_y_counts / total_counts) * 100

# Create a bar plot for the percentages
plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(x=grn_celltype_y_percentage.index, y=grn_celltype_y_percentage.values)

# Set the title and labels of the plot
bar_plot.set_title('Percentage of GRN enrich in Cell Type markers')
bar_plot.set_xlabel('Metrics')
bar_plot.set_ylabel('Percentage of GRN enriched in Cell Type markers')

# Rotate the x labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
